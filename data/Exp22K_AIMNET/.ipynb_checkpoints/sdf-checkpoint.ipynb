```python
import pandas as pd
from rdkit import Chem 

df = pd.read_pickle('NMR18K_FF.pkl')
df
```

```python
NMR18K_FF = []
for idx, row in df.iterrows():
    mol_id = row['Mol_ID']
    mol = row['Optimized_Geometry']
    mol.SetProp('_Name',str(mol_id))
    NMR18K_FF.append(mol)
```

```python
import gzip
with gzip.open("NMR18K_FF.sdf.gz", "wt") as gz:      
    writer = Chem.SDWriter(gz)                       
    for m in NMR18K_FF:
        writer.write(m)
    writer.flush()        
```

```python
import pandas as pd
from rdkit import Chem 

df = pd.read_pickle('Mols_lowest_conf.pkl')
df
```

```python
NMR30K_AIMNET = []
for idx, row in df.iterrows():
    mol_id = row['mol_id']
    mol = row['Mol']
    mol.SetProp('_Name',str(mol_id))
    NMR30K_AIMNET.append(mol)
```

```python
writer = Chem.SDWriter('NMR30K_AIMNET.sdf')
for mol in NMR30K_AIMNET:
    writer.write(mol)
writer.close()
```

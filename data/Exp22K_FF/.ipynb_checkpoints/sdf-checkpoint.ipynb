```python
import pandas as pd
from rdkit import Chem 

df = pd.read_pickle('NMR_22K_AIMNET.pkl')
df
```

```python
NMR22K_AIMNET = []
for idx, row in df.iterrows():
    mol_id = row['Mol_ID']
    mol = row['Optimized_Geometry']
    mol.SetProp('_Name',str(mol_id))
    NMR22K_AIMNET.append(mol)
```

```python
import gzip
with gzip.open("NMR22K_AIMNET.sdf.gz", "wt") as gz:      
    writer = Chem.SDWriter(gz)                       
    for m in NMR22K_AIMNET:
        writer.write(m)
    writer.flush()        
```
